<h2>Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import seaborn as sns
%matplotlib inline
```

<h2>Retrieve data from url and insert into dataframe.

```python
url_trout = "https://www.baseball-reference.com/players/t/troutmi01.shtml"
trout = pd.read_html(url_trout)[1]
trout
```

<h2>Clean Data

```python
#Exclude records that are for minor leagues and also records that show career totals, as I only want major league seasons
trout = trout.query('Lg == "AL" ')
trout
```

```python
#Subset dataset for only features that I want in my analysis
trout = trout[["Year", "G", "PA", "AB", "H"]]
trout
```

```python
#Check datatypes for subsetted data
trout.dtypes
```

```python
#Change 'Year' datatype to int
trout["Year"] = trout["Year"].astype(int)
trout.dtypes
```

```python
#Run corrleation matrix for dataset
trout.corr()
```

```python
#Run corrleation heatmap for dataset
sns.heatmap(trout.corr())
```

```python
trout_test = trout
trout
```

<h2>Train/Test Split

```python
#Create mask to randomly select rows for train/test
msk = np.random.rand(len(trout_test)) <= 0.8
train = trout_test[msk]
test = trout_test[msk]
```

<h2>Start regression model

```python
#Assign training variables
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[["PA"]])
train_y = np.asanyarray(train[["H"]])
regr.fit(train_x, train_y)
#coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
#Evaluate model 
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[["PA"]])
test_y = np.asanyarray(test[["H"]])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```

```python
#See PA values for test set
df_test_x = test_x
df_test_x
```

```python
#Create dataframe for test set and insert PA values 
final_df = pd.DataFrame(df_test_x)
final_df
```

```python
#See Actual Hits values for test set
df_test_y = test_y
df_test_y
```

```python
#Add Actual Hits to dataframe
final_df["Actual Hits"] = df_test_y
final_df
```

```python
#See Predicted Hits values for test set
df_test_y_ = test_y_
df_test_y_
```

```python
#Add Predicted Hits to dataframe
final_df["Predicted Hits"] = df_test_y_
final_df
```

```python
#Change first column name from 0 to PA (Plate Appearances)
final_df.rename(columns = {0:'PA'}, inplace = True)
final_df
```

```python
#Round column Predicted Hits to nearest whole number and show as int
final_df["Predicted Hits"] = round(final_df["Predicted Hits"],0).astype(int)
final_df
```

<h2>Apply model to original dataset

```python
trout_prod = trout
trout_prod
```

```python
#Check datatypes for dataset
trout_prod.dtypes
```

```python
#Run model to predict Hits on original dataset
X_final = trout_prod[["PA"]] 
trout_prod['Preds'] = regr.predict(X_final)
trout_prod
```

```python
#Round predictions to nearest whole number and show as int
trout_prod["Preds"] = round(trout_prod["Preds"], 0).astype(int)
trout_prod
```

```python
#Subset dataset
trout_prod = trout_prod[["Year", "PA", "H", "Preds"]]
```

```python
#Set appropriate columns names
trout_prod.columns = ["Season", "PA", "Actual Hits", "Predicted Hits"]
trout_prod
```

<h2>Visualize Actual Hits vs. Predicted Hits

```python
#Set up figure and axes for plot
fig, ax = plt.subplots(1,1, figsize=(10,8), sharey=True)

#Create plot for regular season stats
ax.scatter(trout_prod["PA"], trout_prod["Actual Hits"], marker="o", color="red")
ax.scatter(trout_prod["PA"], trout_prod["Predicted Hits"], marker="x", color="blue")
ax.set_xlabel("PA", fontweight="bold")
ax.set_ylabel("Hits", fontweight="bold")
ax.set_title("Mike Trout - Actual vs. Predicted Hits", fontweight="bold")
ax.legend(["Actual", "Predicted"], loc="upper left")
```
